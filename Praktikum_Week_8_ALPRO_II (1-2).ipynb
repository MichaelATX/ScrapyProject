```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
response1 = requests.get("https://unair.ac.id/news/")
htmllang1 = response1.text
readhtml1 = BeautifulSoup(htmllang1, "html.parser")

judul1 = []
count1 = 0

for i in readhtml1.find_all("h2"): judul1.append(i.get_text())
for j in readhtml1.find_all("h3"): judul1.append(j.get_text())
for jud1 in judul1:
  count1 += 1
  print(f"{jud1} -> Judul berita ke-{count1}")
```

```python
count = 1

webdata = {"Halaman": [], "Berita": []}
basisrf = "https://unair.ac.id/category/featured/"
datafrm = pd.DataFrame(webdata)

while True:
  link = f"{basisrf}page/{count}/"
  response2 = requests.get(link)

  if response2.status_code == 404:
    break

  readhtml2 = BeautifulSoup(response2.text, "html.parser")
  for news in readhtml2.find_all("h3", class_= "elementor-post__title"):
    judul2 = news.get_text(strip = True)
    datafrm = datafrm.append({"Halaman": count, "Berita": judul2}, ignore_index = True)
  count += 1

datafrm.to_csv("beritaunair.csv", index = False)
```
